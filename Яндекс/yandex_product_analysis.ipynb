## Продуктовый кейс
Допустим, вы работаете в компании, занимающейся социальными сетями, которая только что запустила свой сайт в новом городе. Глядя на еженедельные показатели, вы видите медленное снижение среднего количества комментариев на одного пользователя с января по март в этом городе.

С января по март компания последовательно увеличивала число новых пользователей в городе.

1) Какие уточняющие вопросы стоит задать?
2) Назовите несколько причин, по которым среднее количество комментариев на одного пользователя будет уменьшаться, и на какие показатели вы бы обратили внимание?

### Ответ на 1 вопрос:
Для более полного понимания картины не хватает базовой информации - кто наши пользователи, что это за продукт и как люди с ним взаимодействуют. \
Чтобы получить ответы, я бы задал эти вопросы:

- Кто пользуется этой социальной сетью?
- Как они взаимодействуют друг с другом?
- Были ли какие-либо проблемы с производительностью, которые могли стать причиной проблемы?
- Каковы цели этого конкретного запуска?
- Были ли какие-либо изменения в функциях комментариев за последние недели?

### Ответ на 2 вопрос:

Далее можно сделать предположение, что наш продукт - это социальная сеть, похожая на Facebook, с молодой аудиторией, и целью запуска является расширение пользовательской базы. \
Кроме того, не было никаких проблем с производительностью, и функция комментирования не менялась с момента запуска.ф

Далее вернемся к самому вопросу, нас интересует метрика - `среднее количество комментариев на одного пользователя`. Также как мы знаем, что компания последовательно увеличивала число новых пользоватей, значит мы можем смоделировать примерные данные о пользователях:

`Январь`: 10000 пользователей, 30000 комментариев, 3 комментария на пользователя \
`Февраль`: 20000 пользователей, 50000 комментариев, 2,5 комментария на пользователя \
`Март`: 30000 пользователей, 60000 комментариев, 2 комментария на пользователя

На основе этих смоделированных данных можем сделать 2 вывода:

- Проблема с уменьшением количества комментариев не является результатом сокращения базы пользователей.
- Причина не в технических проблемах платформы.
 
Однако, мы не можем строго ориентироваться на эту метрику, т.к. на самом деле не очень понятно, поможет ли она нам решить этот вопрос.

 

Я бы предложил посмотреть на другие метрика, т.к. `среднее количество комментариев на пользователя` - не самая лучшая метрика, потому что она не дает представления о том, что на самом деле вызывает эту проблему. \
Она не учитывает активных пользователей, которые на самом деле комментируют. Куда более подходящим показателем для исследования были бы метрика удержания `(Retenetion)` пользователей и ежемесячно активных пользователей `(MAU)`.

Можно предположить, что активные пользователи часто комментируют и именно они являются "двигателем" увеличения количества комменатриев из месяца в месяц. \
С другой стороны, новые пользователи могут быть не так вовлечены и не быть такими активными комментаторами. \
Теперь, когда у нас есть гипотеза, было бы неплохо пересобрать наш пазл с учетом новых вводных.

Один из способов решения проблема это - привлечение ежемесячно активных пользователей.

Если количество комментариев растет именно за счет активных пользователей, это помогло бы понять, как мы удерживаем пользователей. Кроме того, это также покажет нам, насколько активны новые пользователи. Менее ли они вовлечены и реже ли они комментируют.

При наличии подохдящих данных, можно было бы сегментировать пользователей по дате их регистрации. Далее визуализировать вовлеченность и увидеть, насколько вовлечены наши "`старички`" и "`новички`".

И если проблема на самом деле состоит в вовлечении новых пользователей, то можно предложить несколько стратегий. Например, протестировать новый функционал регистрации или комментирования, чтобы привлекать больше внимания от пользователей.

### Итоги:
После моих рассуждений я пришел к гипотезе - проблема в вовлечении "`новых`" пользователей и нужно дополнительно изучить "новые" 2 метрики - `Retention(1-го месяца)` и `MAU`.

Дополню свой ответ - если бы причина была не в недостаточной вовлеченности новых пользователей, тогда разумно изучить общую массу активных пользователей. \
Одной из потенциальных причин может быть то, что активные пользователи со временем все меньше комментируют. В таком случае можно сделать вывод, что на самом деле "`старые`" пользователи отваливаются, а вовлеченность "`новых`" потенциально растет.
